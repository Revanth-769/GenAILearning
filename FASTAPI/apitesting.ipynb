```python
import requests
import json

# 1. Define the API endpoint
url = "https://api.euron.one/api/v1/euri/images/generations"

# 2. Set up the headers
#    This includes your authentication token and content type
headers = {
    "Content-Type": "application/json",
    "Authorization": "Bearer euri-7b1470de7f4a72e14de31c442a26dd17ec40acf8d3d0fb8c742ca645abe16cc0"  # <-- Make sure to replace YOUR_API_TOKEN
}

# 3. Define the data payload
#    This is the Python dictionary equivalent of your JSON string
payload = {
    "prompt": "lord rama fighting ravana",
    "model": "black-forest-labs/FLUX.1-schnell",
    "n": 2,
    "size": "1024x1024",
    "quality": "standard",
    "response_format": "url",
    "style": "vivid"
}

# 4. Make the POST request
#    The 'json=payload' argument automatically converts the Python dict
#    to a JSON string and sets the Content-Type header (though we've
#    also set it manually in 'headers' for clarity).
try:
    response = requests.post(url, headers=headers, json=payload)

    # Raise an exception for bad status codes (4xx or 5xx)
    response.raise_for_status() 

    # 5. Print the successful response
    print("Success! Response from server:")
    print(response.json())  # Print the JSON response from the server

except requests.exceptions.HTTPError as errh:
    print(f"Http Error: {errh}")
    print(f"Response content: {errh.response.text}")
except requests.exceptions.ConnectionError as errc:
    print(f"Error Connecting: {errc}")
except requests.exceptions.Timeout as errt:
    print(f"Timeout Error: {errt}")
except requests.exceptions.RequestException as err:
    print(f"Something else went wrong: {err}")
```
